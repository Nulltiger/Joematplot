```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
import seaborn as sns
```

```python
# Make a reference to the city_data.csv file path
city_data = "raw_data/city_data.csv"
city_data_df = pd.read_csv(city_data, encoding="ISO-8859-1")
city_data_df.head()
```

```python
#Make a reference to the ride_data.csv file path
ride_data = "raw_data/ride_data.csv" 
ride_data_df = pd.read_csv(ride_data, encoding="ISO-8859-1")
ride_data_df.head()
```

```python
#Merge the two dataframes into one 
pybermerge_df = pd.merge(ride_data_df, city_data_df, on="city")
pybermerge_df.head()
```

```python
#Create the columns needed inorder to graph
city_df = pybermerge_df.groupby("city")
avg_fare = city_df.mean()["fare"]
total_rides = city_df["ride_id"].count()
drivers = city_df.mean()["driver_count"]
type_of_city = city_data_df.set_index("city")["type"]
```

```python
#New dataframe to hold all graph variables
pyber_df = pd.DataFrame({
    "Average Fare": avg_fare,
    "Total Rides": total_rides,
    "Number of Drivers": drivers,
    "City Type": type_of_city
})
pyber_df.head()
```

```python
#Loc the city inorder to differentiate in scatterplot
rural = pyber_df.loc[pyber_df["City Type"]== "Rural"]
urban = pyber_df.loc[pyber_df["City Type"]== "Urban"]
suburban = pyber_df.loc[pyber_df["City Type"]== "Suburban"]
```

```python
#plot the scatterplots by city type and variables
plt.scatter(rural["Total Rides"],rural["Average Fare"], s = rural["Number of Drivers"]*10, color="green", edgecolors="black", lw = 0.75, alpha = 0.75, label="Rural")
plt.scatter(urban["Total Rides"],urban["Average Fare"], s = urban["Number of Drivers"]*10, color="red",edgecolors="black", lw = 0.75, alpha = 0.75, label="Urban")
plt.scatter(suburban["Total Rides"],suburban["Average Fare"], s = suburban["Number of Drivers"]*10, color="blue",edgecolors="black", lw = 0.75, alpha = 0.75, label="Suburban")


#Assign labels 
plt.suptitle("Pyber Ride Sharing Information")
plt.title("Note: Circle size corresponds to number of drivers")
plt.xlabel("Total Number of Rides by City")
plt.ylabel("Average Fare by City")
plt.legend(loc="best")
plt.grid(lw = 0.5)
plt.show()
```

```python
# Use groupby for fare pie chart
pie1 = pybermerge_df.groupby(["type"])["fare"].sum()
pie1
pie1.index
```

```python

#Create the pie chart#Create  
explode = [0,0,0.1]
colors = ["green", "blue", "red"]
plt.pie(pie1, labels = pie1.index, explode = explode, colors = colors, autopct="%1.1f%%", startangle = 110, shadow = True)
plt.axis("equal")
plt.title("Percent of Total Fares by City Type")
plt.show()
```

```python
#Use groupby for ride pie chart
pie2 = pybermerge_df.groupby(["type"])["ride_id"].count()
pie2.index
```

```python
explode = [0,0,0.1]
colors = ["green", "blue", "red"]
plt.pie(pie2, labels = pie2.index, explode = explode, colors = colors, autopct="%1.1f%%", startangle = 140, shadow = True)
plt.axis("equal")
plt.title("Percent of Total Rides by City")
plt.show()
```

```python
#Create groupby for driver pie chart
pie3 = pybermerge_df.groupby(["type"])["driver_count"].sum()
pie3.index
```

```python
explode = [0,0,0.2]
colors = ["green", "blue", "red"]
plt.pie(pie3, labels = pie3.index, explode = explode, colors = colors, autopct="%1.1f%%", startangle = 140, shadow = True)
plt.axis("equal")
plt.title("Percent of Total Drivers by City Type")
plt.show()
```
